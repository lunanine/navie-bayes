```python

import seaborn as sns
import matplotlib.pyplot as plt

# Bin the weight column into categories
df["weight_bin"] = pd.cut(df["weight"], bins=10)

# Now use countplot on binned data
sns.countplot(x=df["weight_bin"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
